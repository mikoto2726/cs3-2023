```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
csv_in = 'boston_small.csv'
df = pd.read_csv(csv_in,sep=',',skiprows=8,header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_corr = df.corr()
display(df_corr)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_corr,annot=True,fmt='.2f',cmap='bwr')
plt.show()
```

```python
print(df.corrwith(df['MEDV']))
```

```python
plt.scatter(df['MEDV'],df['RM'])
plt.title('corr:{:.2f}'.format(df_corr.at['MEDV','RM']))
plt.xlabel('MEDV')
plt.ylabel('RM')
plt.show()
```
