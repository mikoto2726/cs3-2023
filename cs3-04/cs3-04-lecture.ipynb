#### Cleansing of wholesale_customers data

#### Read CSV file

```python
csv_in = 'wholesale_customers-with-errors.csv'
df = pd.read_csv(csv_in, sep=',', comment='#', header=0)
#df = pd.read_csv(csv_in, sep=',', skiprows=13, header=0)
df['Channel'] = df['Channel'].astype('category')
print(df.shape)
print(df.info())
display(df.head())
```

#### Check duplicated data

```python
display(df[ df.duplicated(keep=False) ])
```

#### Delete duplicated data

```python
df = df.drop_duplicates().reset_index(drop=True)
print(df.shape)
display(df.head())
```

#### Check missing values

```python
print(df.isnull().sum())
```

```python
display(df[df.isnull().any(axis=1)])
```

#### Fill missing values

```python
mu = df.mean(numeric_only=True)
print(mu)
```

```python
df_filled = df.fillna( mu )
display(df_filled.head())
```

#### Delete data with missing values

```python
df_del = df.dropna()
display(df_del.head())
```

```python
df_del2 = df.dropna().reset_index(drop=True)
display(df_del2.head())
```

```python
df = df.dropna().reset_index(drop=True)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df['Milk'] = df['Milk'].astype('int')
df['Frozen'] = df['Frozen'].astype('int')
print(df.info())
display(df.head())
```

#### Check notation fluctuation

```python
print(df['Region'].value_counts())
```

```python
df['Region'] = df['Region'].replace('other', 'Other')
df['Region'] = df['Region'].replace(['lisbon', 'Lisbn'], 'Lisbon')
df['Region'] = df['Region'].replace({'porto': 'Oporto', 
                                     'Porto': 'Oporto',
                                     'Otheer': 'Other',
                                     })
print(df['Region'].value_counts())
```

#### (Adv) Replace using map method

```python
print(df['Fresh'].head())
ser_fresh = df['Fresh'].map( lambda x: 10000 if x > 10000 else x )
display(ser_fresh.head())
```

```python
csv_out = 'wholesale_customers-cleaned.csv'
df.to_csv(csv_out, index=False, encoding='utf-8-sig')
```
