```python
import numpy as nd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_in = 'hakusan_elec.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=1, header=0)
print(df.shape)
print(df.info())
display(df.head())
display(df.tail())
```

```python
#datetime型に変換するためにはpd.to_datetime関数を使う
df['Date']=pd.to_datetime(df['Date'])
print(df.info())
```

```python
plt.plot(df['Date'], df['bldg_1'])
plt.xlabel('Date')
plt.ylabel('Electricity Power Consumption [kWh]')
plt.show()
```

```python
#plt.plot(x軸用データ,y軸用データ,[フォーマット文字列],オプション引数)
#データは2つの1次元データ(第1,2引数)
#フォーマット文字列:'マーカー指定線指定色指定'(指定しなくてもよい)
#マーカー指定:'o':丸, 's':四角, 'x':x印, ',':点,
#線指定:'-':実線, '--':破線, ':':点線
#色指定:'r':赤, 'g':緑, 'b':青, 'y':黄
#例:'x-r':xマーカーと赤い実線
#'--b':マーカー無し青い実線
#label='凡例用ラベル'
#linewidth=N 線の太さ指定
```

```python
#日付列をindexに指定する
df=df.set_index('Date')
display(df.head())
```

```python
#datetime型のindexから日付などの情報を取得するためのメソッドが用意されている
#以下先頭の3データだけを表示
#print(df.index[:3])
#print(df.index.year[:3])
#month
#day
#date 年月日のndarray
#dayofyear　1年の何日目か
#dayofweek　曜日番号0~7
#quarter　四半期番号
```

```python
display(df.head(10))
display(df.tail())
```

```python
df_rolling7=df.rolling(7).mean() #窓幅7の移動平均を求める
display(df_rolling7.head(10))
display(df_rolling7.tail())
```

```python
df_rolling7=df.rolling(7).mean()
display(df_rolling7.head(10))
display(df_rolling7.tail())
```

```python
df_rolling7d=df.rolling('7d').mean()
display(df_rolling7d.head(10))
display(df_rolling7.tail())
```

```python
plt.plot(df.index, df['bldg_1'], label='bldg_1')
plt.plot(df_rolling7d.index, df_rolling7d['bldg_1'], label='rolling7d')
plt.xlabel('Date')
plt.ylabel('Electricity Power Consumption [kWh]')
plt.legend()
plt.show()
```

```python
start_date=pd.to_datetime('2018-04-01')
end_date=pd.to_datetime('2018-06-30')
plt.plot(df.index, df['bldg_1'], label='bldg_1')
plt.plot(df_rolling7d.index, df_rolling7d['bldg_1'],label='rolling7d')
plt.xlim(start_date, end_date)
plt.xlabel('Date')
plt.ylabel('Electricity Power Consumption [kWh]')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
