```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

csv_in = '../cs3-02/running_tab.csv'
df1 = pd.read_csv(csv_in, sep='\t', skiprows=0, header=0)

print(df1.shape)
print(df1.info())
print(df1.head())
```

```python
print(df1['Univ'].value_counts())
```

```python
display(df1['Univ'].describe())
```

```python
ave_gra = df1['Grade'].mean()
print(ave_gra)
std_gra = df1['Grade'].std()
print(std_gra)
```

```python
ave_time = df1['Time'].mean()
print(ave_time)
std_time = df1['Time'].std()
print(std_time)
```

```python
plt.boxplot(df1['Time'])
plt.ylabel('Time[sec.]')
plt.show()
```

```python
plt.hist(df1['Time'])
plt.xlabel('Time[sec.]')
plt.ylabel('Frequency')
plt.show()
```
