```python
import os
os.environ['OMP_NUM_THREADS'] = '1'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
csv_in = 'winequality-red.csv'
df = pd.read_csv(csv_in, sep=';', skiprows=0, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
display(df.describe())
```

```python
X_scaled = scale(df)
```

```python
print(X_scaled.mean(axis=0))
print(X_scaled.std(ddof=0, axis=0))
```

```python
df_scaled = pd.DataFrame(X_scaled, columns=df.columns)
display(df_scaled.head())
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, n_init=10)
    km.fit(df_scaled)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
n_cls = 4
km = KMeans(n_clusters=n_cls, n_init=10, random_state=5)
cls = km.fit_predict(df_scaled)
print(cls)
```

```python
df['cluster_no'] = cls
display(df.head())
```

```python
print(df['cluster_no'].value_counts())
```

```python
cols = list(df.columns)
for c in cols:
    if c == 'cluster_no': continue
    dat = []
    for i in range(n_cls):
        df_cls = df[ df['cluster_no']==i ]
        dat.append(df_cls[c])
    plt.boxplot(dat, labels=range(n_cls))
    plt.xlabel('Cluster No.')
    plt.ylabel(c)
    plt.show()
```
