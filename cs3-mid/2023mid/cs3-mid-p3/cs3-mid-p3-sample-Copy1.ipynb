```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_in = 'corona-data.csv'
df=pd.read_csv(csv_in, skiprows=0, header=0)
```

```python
wday_list = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
weekend = wday_list[-2:]
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df=df.set_index('date')
df_week=df.resample('W').mean()
display(df_week)
```

```python
df['day_of_week']=df.index.dayofweek
```

```python
df_wday_ave=df.groupby('day_of_week').mean()
```

```python
display(df_wday_ave)
```

```python
wday_list = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']  
weekend = wday_list[-2:]
plt.pie(df_wday_ave['positive_cases'][-2:], labels=weekend)
plt.show()
```
