```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
```

```python
img_in = 'cat_sample.jpg'
img = Image.open(img_in)
```

```python
print(img.format, img.size,
      img.width, img.height, img.mode)
```

```python
plt.imshow(img)
plt.show()
```

```python
data=np.array(img)
print(data.shape)
```

```python
print(img.getpixel((200, 800)))
```

```python
img_small = img.resize((30,50))
plt.imshow(img_small)
plt.show()
```

```python
img_rot = img.rotate(60)
plt.imshow(img_rot)
plt.show()
```

```python
img_blur = img.filter(ImageFilter.GaussianBlur(radius=10))
plt.imshow(img_blur)
plt.show()
```

```python
img_crop = img.crop((300, 400, 600, 700))
plt.imshow(img_crop)
plt.show()
```

```python
img_gray = img.convert('L')  # gray scale
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
img_out = 'cat_gray.jpg'
img_gray.save(img_out, quality=95)
```
